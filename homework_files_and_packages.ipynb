```python
import json
f = open('purchase_log.txt', encoding= 'utf-8')
purchases = {}
for i, line in enumerate(f):
    line = line.strip()
    users = json.loads(line)
    user_id = users['user_id']
    category = users['category']
    purchases[user_id] = category
    # print (purchases)  #оставлено для контроля первой задачи
    # if i>5:
    #     break
```

```python
import csv
csvvisit = open('visit_log.csv', newline='', encoding='utf-8')
csvfunnel = open('funnel.csv', 'w', newline='', encoding='utf-8')
writer = csv.writer(csvfunnel)
# writer.writerow(['user_id', 'source', 'category'])
for row in csv.reader(csvvisit):
    user_id = row[0]
    source = row[1]
    if user_id in purchases:
        category = purchases[user_id]
        writer.writerow([user_id, source, category]) 
```

```python
f.close()
csvvisit.close()
csvfunnel.close()
```
