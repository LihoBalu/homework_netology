```python
import pandas as pd
```

```python
tags_df = pd.read_csv('tags.csv')
tags_df.head()
```

```python
grouped_tags = tags_df.groupby('movieId')['tag'].apply(lambda x:', '.join(x)).reset_index()

grouped_tags.head()
```

```python
rating_df = pd.read_csv('ratings.csv')
rating_grouped = rating_df.groupby('movieId')['rating'].mean().reset_index()
rating_grouped.head()
```

```python
rating_tag = rating_grouped.merge(grouped_tags, on ='movieId', how='left')
rating_tag.head()
```

```python
movies_df = pd.read_csv('movies.csv')
movies_df.head()
```

```python
movies_rating_tag = movies_df.merge(rating_tag, on ='movieId', how='left')
movies_rating_tag.head(15)
```

```python
movies_rating_tag.to_excel('testing_write.xlsx', sheet_name='Получилось')
```
