```python
import pandas as pd
df_ratings = pd.read_csv('ratings.csv')
df_movies = pd.read_csv('movies.csv')
movies_stats = df_ratings.merge(df_movies, on='movieId')
start_year = 1950
end_year = 2010
years = [year for year in range(start_year, end_year + 1)]
def production_year(title):
    for year in years:
        if str(year) in title:
            return year
    return 1900
movies_stats['year'] = movies_stats['title'].apply(production_year)
average_rating_by_year = movies_stats.groupby('year')['rating'].mean()
sorted_average_rating = average_rating_by_year.sort_values(ascending=False)
print(sorted_average_rating)
```
