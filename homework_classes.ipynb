```python
import requests
import json

def get_max_exchange_rate_currency():
    response = requests.get("https://www.cbr-xml-daily.ru/daily_json.js")
    if response.status_code == 200:    # проверка на успешный ответ от сайта
        data = json.loads(response.text)
        currencies = data["Valute"]
        max_currency = max(currencies.values(), key=lambda x: x["Value"])
        return max_currency["Name"]
    else:
        return "Не удаётся получить данные"

max_currency_name = get_max_exchange_rate_currency()
print(f"{max_currency_name}")
```
