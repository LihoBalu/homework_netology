# Задача 1

```python
import pandas as pd
df = pd.read_csv('ratings.csv')
def classification_films(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    if 2 <=rating <= 4:
        return 'средний рейтинг'
    if rating >= 4.5:
        return 'высокий рейтинг'
df['class'] = df['rating'].apply(classification_films)
df.head()
```

# Задача 2

```python
import pandas as pd
search_df = pd.read_csv('keywords.csv')
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def search_regions(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if city.lower() in keyword.lower():
                return region
            return 'undefined'
search_df['region'] = search_df['keyword'].apply(search_regions)
search_df.head()
```
