# Задача 1

```python
import pandas as pd
df_ratings = pd.read_csv('ratings.csv')
df_movies = pd.read_csv('movies.csv')
movies_merge = df_ratings.merge(df_movies, on='movieId')
reiting_five = movies_merge[movies_merge['rating'] == 5.0]
best_film = reiting_five['title'].mode()
print(best_film)
```

```python
data = pd.read_csv('power.csv')
baltica = data[(data['country']=='Estonia') | (data['country']=='Latvia') | (data['country']=='Lithuania')]
baltica_cat = baltica[baltica['category'].isin([4, 12, 21])]
baltica_cat_year = baltica_cat[(baltica_cat['year'] >= 2005) & (baltica_cat['year']<=2010)]
result = baltica_cat_year[baltica_cat_year['quantity']>0].sum()
result['quantity']
```
